### Q.13. Show Top 10 Countries with the Most Content.

```python
# Count the number of shows by country
import matplotlib.pyplot as plt
import seaborn as sns
country_counts = data['Country'].value_counts().head(10)

# Plot the top 10 countries
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.values, y=country_counts.index, palette='plasma')
plt.title('Top 10 Countries with the Most Content on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```
